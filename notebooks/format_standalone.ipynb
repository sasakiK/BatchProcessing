```python
import os,datetime, collections
from pyspark.sql import SQLContext, Row
from pyspark.sql.types import *
from pyspark import SparkConf, SparkContext
from pyspark.sql import functions as sf
import pandas as pd
```

```python
%%timeit
schema=StructType([StructField("ID", StringType()),
                   StructField("V0", IntegerType()),
                   StructField("V1", IntegerType()),
                   StructField("V2", IntegerType()),
                   StructField("V3", IntegerType()),
                   StructField("V4", IntegerType()),
                   StructField("V5", IntegerType()),
                   StructField("V6", IntegerType()),
                   StructField("V7", IntegerType()),
                   StructField("V8", IntegerType()),
                   StructField("V9", IntegerType())])

whole_data = spark.read.csv('file:///etc/hadoop/hadoop_sample_10g.csv', schema=schema)
whole_data.show(5)
# show partitions
print("Partition is ... {}".format(whole_data.rdd.getNumPartitions()))
# show schema
print("================\n {}".format(whole_data.printSchema()))

# reformat one columns
whole_rdd_formatted = whole_data.withColumn('ID', sf.concat(sf.col('ID'),sf.lit('-00')))
whole_rdd_formatted.show()
```
